```python
#importing the python libraries
# Data Handling
import numpy as np
import pandas as pd

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical Analysis
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

## Data Importing and Preprocessing

```python
# importing csv file
csv_filename = "../data/Kilifi_data.csv"
```

```python
# reading csv file using pandas and converting into a DataFrame
df_Mvolkensii = pd.read_csv(csv_filename)
df_Mvolkensii
```

```python
#checking column datatypes
df_Mvolkensii.dtypes
```

```python
# converting Site column to categorical data type
df_Mvolkensii['Site'] = df_Mvolkensii['Site'].astype('category')
```

```python
#checking column datatypes
df_Mvolkensii.dtypes
```

```python
# checking and counting column missing values
df_Mvolkensii.isna().sum()
```

# Diameter at Breast Height(DBH) Analysis

## Descriptive Statistics

```python
# descriptives statistics for dbh
df_Mvolkensii.groupby('Site', observed=True)['DBH'].describe()
```

## Data visualization

### DBH Boxplot

```python
# Defining color palette
palette_diameter = ["red", "yellow", "orange"] 

# boxplot for DBH
fig_dbh = plt.figure(figsize=(8,6)) # Create a figure of 8x6 inches using plt.figure
sns.boxplot(x='Site', y='DBH', 
            data=df_Mvolkensii, 
            hue='Site', 
            palette=palette_diameter, 
            legend=False)  #Create a box plot using Seaborn
plt.xlabel("Site",fontsize=14) # Set the x-axis label
plt.ylabel("DBH (cm)",fontsize=14) # Set the y-axis label
plt.title("Boxplot of DBH by Site", fontsize=16, fontweight="bold") # Set the title
plt.show( fig_dbh ) # Show the figure in the screen
plt.close( fig_dbh ) # Close the figure 
```

## Statistical Analysis

### Levene's Test

```python
# Levene's Test for homogeneity of variances for DBH
f_value, p_value = stats.levene(*[df_Mvolkensii[df_Mvolkensii['Site'] == s]['DBH'] for s in df_Mvolkensii['Site'].unique()])
print(f"Levene’s test for DBH: F value={f_value:.4f}, P value={p_value:.4f}")
```

### One Way Anova

```python
# ANOVA for DBH
model_dbh = smf.ols('DBH ~ Site', data=df_Mvolkensii).fit()# ANOVA model
anova_table = sm.stats.anova_lm(model_dbh, typ=2)
print("\nANOVA results for DBH:\n", anova_table)
```

### Tukey’s HSD test

```python
# Tukey’s HSD test for DBH
tukey_dbh = pairwise_tukeyhsd(endog=df_Mvolkensii['DBH'], groups=df_Mvolkensii['Site'], alpha=0.05)
print("\nTukey HSD results for DBH:\n", tukey_dbh)
```

# Height Analysis

## Descriptive Statistics

```python
# descriptives statistics for height
df_Mvolkensii.groupby('Site', observed=True)['Height'].describe()
```

## Data Visualization

### Height Boxplot

```python
# boxplot for Height
fig_height = plt.figure(figsize=(8,6)) # Create a figure of 8x6 inches using plt.figure
sns.boxplot(x='Site', y='Height', 
            data=df_Mvolkensii, 
            hue='Site', 
            palette="Set2", 
            legend=False)  # Create a box plot using Seaborn
plt.xlabel("Site",fontsize=14,) # Set the x-axis label
plt.ylabel("Height (m)",fontsize=14,) # Set the y-axis label
plt.title("Boxplot of Height by Site", fontsize=16, fontweight="bold") # Set the title
plt.show( fig_height ) # Show the figure in the screen
plt.close( fig_height ) # Close the figure 
```

## Statistical Analysis

### Levene's Test

```python
# Levene's Test for homogeneity of variances for height
f_value, p_value = stats.levene(*[df_Mvolkensii[df_Mvolkensii['Site'] == s]['Height'] for s in df_Mvolkensii['Site'].unique()])
print(f"Levene’s test for Height: F value={f_value:.4f}, P value={p_value:.4f}")
```

### One Way Anova

```python
# One Way ANOVA for height
model_height = smf.ols('Height ~ Site', data=df_Mvolkensii).fit()# ANOVA model
anova_table = sm.stats.anova_lm(model_height, typ=2)
print("\nANOVA results for Height:\n", anova_table)
```

### Tukey’s HSD test

```python
# Tukey’s HSD test for Height
tukey_height = pairwise_tukeyhsd(endog=df_Mvolkensii['Height'], groups=df_Mvolkensii['Site'], alpha=0.05)
print("\nTukey HSD results for Height:\n", tukey_height)
```
